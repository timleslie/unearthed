```python
columns = {
    "3311WI671.PV" : "Spewput",
    "3311HS181A.PV" : "Throughput",
}
```

```python
data.rename(columns=columns, inplace=True)
```

```python
SPEW_EVENTS = False
POST_SPEW = False
PRE_SPEW = False
PREDICT_POINT = False
PREDICT_HISTORY = False
LOTS_OF_DATA = False
LOTS_OF_DATA_FILTERED = False
RESULTS = False
```

```python
def DoThePlot(window_size="long"):
    if window_size == "long":
        slice = data.loc[pd.Timestamp('2016-07-28 06:00') : pd.Timestamp('2016-07-29 06:00')]
    elif window_size == "short":
        slice = data.loc[pd.Timestamp('2016-07-28 21:00') : pd.Timestamp('2016-07-28 22:45')]
    else:
        slice = data.loc[pd.Timestamp('2016-07-28 21:20') : pd.Timestamp('2016-07-28 22:00')]

    spewevent = slice.SPEW_EVENT
    prespew = slice.PRE_SPEW
    afterspew = slice.POST_SPEW

    slice[["Spewput"]].plot()

    if SPEW_EVENTS:
        for time, spew in spewevent[spewevent == True].iteritems():
            plt.axvline(time, color='r')

    if POST_SPEW:
        for begin, end in zip(
            afterspew[(afterspew.diff()!= 0) & afterspew].index,
            afterspew[(afterspew.diff()!= 0) & ~afterspew].iloc[1:].index
        ):
            plt.axvspan(begin, end, color='r', alpha = 0.3)

    if PRE_SPEW:
        for begin, end in zip(
            prespew[(prespew.diff()!= 0) & prespew].index,
            prespew[(prespew.diff()!= 0) & ~prespew].iloc[1:].index
        ):
            plt.axvspan(begin, end, color='g', alpha = 0.3)

    HISTORY_PERIOD = pd.Timedelta('00:05:00')
            
    for predict_point in prespew[(prespew.diff()!= 0) & prespew].index:
        if PREDICT_POINT:
            plt.axvline(predict_point, color='g')
        if PREDICT_HISTORY:
            plt.axvspan(predict_point - HISTORY_PERIOD, predict_point, color='y', alpha = 0.3)
            
    if LOTS_OF_DATA:
        normalized = (slice - slice.mean()) / (slice.max() - slice.min())
        if LOTS_OF_DATA_FILTERED:
            first_prediction = prespew[(prespew.diff()!= 0) & prespew].index[0]
            normalized = normalized.loc[first_prediction - HISTORY_PERIOD: first_prediction]
            
        normalized.plot(legend=False, linewidth=0.3, secondary_y=True, ax=plt.gca())
        plt.gca().yaxis.set_major_locator(plt.NullLocator())
    print(sum(2500.0 - slice.Throughput[afterspew])/len(slice.Throughput[afterspew]))
#     if RESULTS:
#         ax = plt.gca()
#         results_slice.plot(secondary_y=True, ax=ax)
```
